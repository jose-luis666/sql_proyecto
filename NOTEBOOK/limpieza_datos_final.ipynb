## Tabla Film ##

```python
df_film = pd.read_csv('./data/film.csv')
```

```python
df_film.head()
```

```python
df_film.info()
```

```python
# Ahora, para incluir a los actores de la tabla Actors en nuestra tabla Films, definiremos una función.

# Después de darle muchas vueltas, confieso que acudí al proyecto de Carlos para resolver esta parte:

# El argumento de entrada será el título de una película
# Esta función realizará un bucle "for" que recorrerá el índice de "df_old" y comparará el título de cada fila con el nombre de la película de entrada. 
# Si la fila corresponde a la película buscada, se guardará el nombre del actor/actriz en una lista "actores_actrices". 
# Finalmente, la función unirá todos los nombres de los actores en un string separados por coma, usando la función "join".

df_film['actors'] = '' 
def actors(pelicula): 
    actores_actrices = []
    for i in df_old.index:
        if df_old.title[i] == pelicula:
            name = df_old.name_surname[i]
            actores_actrices.append(name)
    return ', '.join(actores_actrices)
```

```python
df_film
```

```python
# Ahora utilizamos un nuevo bucle for para introducir los actores en cada una de sus filas

for i in df_film.index:
    df_film['actors'][i] = actors(df_film.title[i])
```

```python
df_film
```

```python
# Relleno los elementos vacíos con valores nulos, para luego convertirlos en "unknown actors"
import numpy as np

df_film['actors'] = df_film['actors'].replace('', np.nan)
df_film['actors'] = df_film['actors'].fillna('unknown actors')
```

```python
# Hago algunos cambios en los nombres de las columnas para que no haya confusiones con las palabras seleccionadas de SQL

df_film = df_film.rename(columns={'description': 'description_', 'length': 'length_'})
```
